```python
## Importing Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
## Question :- From the URL:'https://books.toscrape.com/', use web-scrapping and store the Title, Link, Price, Quantity of books
## Csv file.
```

```python
## Ans:-
```

```python
base_url = 'https://books.toscrape.com/catalogue/'
current_page = 'http://books.toscrape.com/catalogue/page-1.html'

response = requests.get(current_page)
book_details = []
while response.status_code == 200:
    
    data = BeautifulSoup(response.text,'html.parser')
    books = data.find_all(class_='product_pod')
    
    for book in books:
        url = base_url + book.h3.a['href']
        res = requests.get(url)
        data1 = BeautifulSoup(res.text,'html.parser')
        
        title = data1.h1.string.strip()
        price = data1.find(class_='price_color').string
        qty = data1.find(class_='instock availability')
        qty = qty.contents[-1].strip()
        qty = int(re.search('\d+',qty).group())
        price = float(re.search('[\d.]+',price).group())
        
        book_details.append([title,url,price,qty])
        
    next_page = data.find(class_='next')
    if next_page is None:
        break
    next_page_url = base_url + next_page.a['href']
    
    current_page = next_page_url
    response = requests.get(current_page)
```

```python
df = pd.DataFrame(book_details, columns = ['Title','Link','Price','Quantity in Stock'])
df.to_csv('books.csv',index=False)
```
